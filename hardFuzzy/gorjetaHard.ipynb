```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x_comida = np.arange(0, 11, 1)
x_servico = np.arange(0, 11, 1)
x_gorjeta = np.arange(0, 21, 1)
```

```python
x_comida, x_servico, x_gorjeta
```

```python
y_comida_ruim = fuzz.trimf(x_comida, [0, 0 , 5])
y_comida_ruim
```

```python
y_comida_boa = fuzz.trimf(x_comida, [0 , 5, 10])
y_comida_boa
```

```python
y_comida_saborosa = fuzz.trimf(x_comida, [5 , 10, 10])
y_comida_saborosa
```

```python
plt.plot(x_comida, y_comida_ruim, 'bo', label = 'ruim')
plt.plot(x_comida, y_comida_ruim)
plt.legend()
```

```python
plt.plot(x_comida, y_comida_boa, 'bo', label = 'boa')
plt.plot(x_comida, y_comida_boa)
plt.legend()
```

```python
plt.plot(x_comida, y_comida_saborosa, 'bo', label = 'saborosa')
plt.plot(x_comida, y_comida_saborosa)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x_comida, y_comida_ruim, 'b', label = 'ruim')
ax.plot(x_comida, y_comida_boa, 'g', label = 'boa')
ax.plot(x_comida, y_comida_saborosa, 'r', label = 'saborosa')
ax.legend
```

```python
y_servico_ruim = fuzz.trimf(x_servico, [0, 0 , 5])
y_servico_ruim
```

```python
y_servico_aceitavel = fuzz.trimf(x_servico, [0, 5 , 10])
y_servico_aceitavel
```

```python
y_servico_otimo = fuzz.trimf(x_servico, [5, 10 , 10])
y_servico_otimo
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x_servico, y_servico_ruim, 'b', label = 'ruim')
ax.plot(x_servico, y_servico_aceitavel, 'g', label = 'aceitavel')
ax.plot(x_servico, y_servico_otimo, 'r', label = 'otimo')
ax.legend
```

```python
y_gorjeta_baixa = fuzz.trimf(x_gorjeta, [0, 0 , 10])
y_gorjeta_media = fuzz.trimf(x_gorjeta, [0, 10 , 20])
y_gorjeta_alta = fuzz.trimf(x_gorjeta, [10, 20 , 20])


```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x_gorjeta, y_gorjeta_baixa, 'b', label = 'baixa')
ax.plot(x_gorjeta, y_gorjeta_media, 'g', label = 'media')
ax.plot(x_gorjeta, y_gorjeta_alta, 'r', label = 'alta')
ax.set_title('Gorjeta')
ax.legend
```

```python
input_comida = 8.0
input_servico = 6.5
```

```python
comida_nivel_ruim = fuzz.interp_membership(x_comida, y_comida_ruim, input_comida)
comida_nivel_boa = fuzz.interp_membership(x_comida, y_comida_boa, input_comida)
comida_nivel_saborosa = fuzz.interp_membership(x_comida, y_comida_saborosa, input_comida)
```

```python
servico_nivel_ruim = fuzz.interp_membership(x_servico, y_servico_ruim, input_servico)
servico_nivel_aceitavel = fuzz.interp_membership(x_servico, y_servico_aceitavel, input_servico)
servico_nivel_otimo = fuzz.interp_membership(x_servico, y_servico_otimo, input_servico)
```

```python
ativacao_regra1 = np.fmax(comida_nivel_ruim, servico_nivel_ruim)
ativacao_gorjeta_baixa = np.fmin(ativacao_regra1, y_gorjeta_baixa)
```

```python
ativacao_gorjeta_media = np.fmin(servico_nivel_aceitavel, y_gorjeta_media)
```

```python
ativacao_regra3 = np.fmax(comida_nivel_saborosa, servico_nivel_otimo)
ativacao_gorjeta_alta = np.fmin(ativacao_regra3, y_gorjeta_alta)
```

```python
x_gorjeta0 = np.zeros_like(x_gorjeta)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.fill_between(x_gorjeta, x_gorjeta0, ativacao_gorjeta_baixa, facecolor = 'b')
ax.plot(x_gorjeta, y_gorjeta_baixa, 'b')
ax.fill_between(x_gorjeta, x_gorjeta0, ativacao_gorjeta_media, facecolor = 'g')
ax.plot(x_gorjeta, y_gorjeta_media, 'g')
ax.fill_between(x_gorjeta, x_gorjeta0, ativacao_gorjeta_alta, facecolor = 'r')
ax.plot(x_gorjeta, y_gorjeta_alta, 'r')
```

```python
controle = np.fmax(ativacao_gorjeta_alta, np.fmax(ativacao_gorjeta_media, ativacao_gorjeta_baixa))
gorjeta = fuzz.defuzz(x_gorjeta, controle, 'centroid')
```

```python
gorjeta_ativacao = fuzz.interp_membership(x_gorjeta, controle, gorjeta)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x_gorjeta, y_gorjeta_baixa, 'b', label = 'baixa')
ax.plot(x_gorjeta, y_gorjeta_media, 'g', label = 'media')
ax.plot(x_gorjeta, y_gorjeta_alta, 'r', label = 'alta')
ax.set_title('Gorjeta')
ax.fill_between(x_gorjeta, x_gorjeta0, controle, facecolor = 'purple')
ax.plot([gorjeta, gorjeta], [0, gorjeta_ativacao], 'black')
ax.legend
```
